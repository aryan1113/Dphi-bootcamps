# **Task2 : Visualization of results using Matplotlib library**

* Display the data of top 5 east asian countries based on 'Generosity'
* Build a plot (line plot) that shows the variation of 'Ladder score' among the above 5 east asian countries (based on generosity)
* Build a plot that shows the variation of 'Ladder score' among the 5 south east asian countries
* Create a dataframe object : df_2021 with the following countries: 'China', 'Nepal', 'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan'. Now, build a scatter plot to show the relation these countries vs their 'Logged GDP per capita'

```python
# Display the data of top 5 east asian countries based on 'Generosity'

df_east_asia=df1 [ df1 ['Regional indicator']=='East Asia']    #All East Asian countries

df_east_asia_1=df_east_asia.sort_values(by=['Generosity'])[:5]   #Top 5 countries based on column score 'Generosity'

df_east_asia_1
```

```python
# Build a plot (line plot) that shows the variation of 'Ladder score' among the above 5 east asian countries (based on generosity)

import matplotlib.pyplot as plt

y=[1,2,3,4,5]
temp=[i for i in df_east_asia_1['Ladder score']]

plt.plot(y,df_east_asia_1['Ladder score'],marker='X')

countries=[z for z in df_east_asia_1['Country name']]

for i,countries in enumerate(countries):
    plt.annotate( countries , ( y[i] ,temp[i] ) )

plt.ylabel('Ladder score ')
plt.show()
```

```python
# Build a plot that shows the variation of 'Ladder score' among the 5 south east asian countries


df_south_east_asia=df1 [ df1 ['Regional indicator']=='Southeast Asia']
# df_south_east_asia       # To see all countries which are in Southeast Asia
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(9,5))
y=[1,2,3,4,5,6,7,8,9]
temp=[i for i in df_south_east_asia['Ladder score']]

plt.plot(y,df_south_east_asia['Ladder score'])

countries=[z for z in df_south_east_asia['Country name']]

for i,countries in enumerate(countries):
    plt.annotate( countries , ( y[i] ,temp[i] ) )

plt.ylabel('Ladder score ')
plt.title('Ladder score of Southeast Asian countries')

plt.show()
```

```python
'''Create a dataframe object : df_2021 with the following countries: 
'China', 'Nepal', 'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan'. 
Now, build a scatter plot to show the relation these countries vs their 'Logged GDP per capita'''


df_2021=df1 [ df1 ['Country name']=='China']
df_2021=df_2021.append(df1 [ df1 ['Country name']=='Nepal'])
df_2021=df_2021.append(df1 [ df1 ['Country name']=='Bangladesh'])
df_2021=df_2021.append(df1 [ df1 ['Country name']=='Pakistan'])
df_2021=df_2021.append(df1 [ df1 ['Country name']=='Myanmar'])
df_2021=df_2021.append(df1 [ df1 ['Country name']=='India'])
df_2021=df_2021.append(df1 [ df1 ['Country name']=='Afghanistan'])

y=[1,2,3,4,5,6,7]
score_countries=[pqrs for pqrs in df_2021['Ladder score']]
countries=['China', 'Nepal', 'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan']
plt.scatter(y,df_2021['Ladder score'])

for i,countries in enumerate(countries):
    plt.annotate( countries , ( y[i] ,score_countries[i] ) )
    
plt.title("Ladder score on y-axis")
plt.show()
```
